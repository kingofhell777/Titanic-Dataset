```python
import pandas as pd
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
```

```python
df = pd.read_csv (r'C:\Users\Aditya Sharma\Pictures\Desktop\archive\tested.csv')
ages = df['Age'].dropna()

```

```python
plt.figure(figsize=(8,5))
sns.histplot(ages, bins=30, kde=True, color="Yellow")
plt.title("Distribution of Passenger Ages")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
mean_age = ages.mean()
median_age = ages.median()
mode_age = ages.mode()[0]
trimmed_mean_age = stats.trim_mean(ages, 0.1)
skewness_age = stats.skew(ages)

print ("Mean:" , mean_age)
print("Median:", median_age)
print("Mode:", mode_age)
print("10% Trimmed Mean:" , trimmed_mean_age)
print ("skewness:", skewness_age)
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x=ages, color="magenta")
plt.title("Boxplot of Passenger Ages")
plt.xlabel("Age")
plt.show()
```

```python
print("The age distribution of titanic passengers is right skewed , with few individuals clustered within the younger ages and few much older passengers streching the tail.In such skewed distributions the mean is pulled by extreme values , while the median remains more representative of the typical passenger.")
```
